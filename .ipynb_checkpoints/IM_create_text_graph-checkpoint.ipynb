```python
def create_dictionary(folder, node_set, layer_number):
    '''
    function that returns a dictionary of the format {node: (relationship, neighbouring node)}
    for up to the degree specified by the layer_number
    starting from a given node_set
    based on the graph given in the folder
    '''
    start = time.time()
    final_dictionary = defaultdict(set)
    
    for layer in range(layer_number):
        
        for root, dirs, files in (os.walk(folder)):
            #print(files)
            for file in tqdm_notebook(files):
                
                #if file.endswith(".csv"):
                
                    with open(str(folder)+str(file), 'rt', encoding = 'utf-8') as f:
                        reader = csv.DictReader(f)
                        for line in (reader):

                            if (line['URI']) in node_set:

                                #final_dictionary[line['URI']] = [tup for tup in (list(line.items())) if tup[1] != 'NULL'] 
                                final_dictionary[line['URI']] = line
                                
        node_set = create_set_of_nodes(final_dictionary)
    end = time.time()
    print('Time taken:', end - start)
    return final_dictionary
```

```python
def create_set_of_nodes(dic):
    '''
    function that returns a set of nodes
    that appear in a dictionary created by the create_dictionary() function
    '''
    new_set = set()
    for node in (dic.values()):
        
        new_list = ([tup[1].replace('{','').replace('}','').split('|') for tup in node if 'db' in tup[1]])
        
        new_set.update([node for tup in new_list for node in tup])
        
    return new_set
```

```python
def create_pairs(graph, annotation):
    '''
    function that takes a graph and annotations
    and returns all combinations of the nodes
    graph argument is needed because of the select_related_nodes() function
    '''
    
    all_annotations = []
    if type(annotation) != list:
        annotation = literal_eval(annotation)
    for sentence in ((annotation)):
        if type(sentence) == list:
            for node in (sentence):  
                all_annotations.append(node)
        else:
            all_annotations.append(sentence)
            
    annotation_pairs = itertools.combinations([node['URI'] for node in all_annotations], 2)
    return set(annotation_pairs)
```

```python
def select_related_nodes(graph, annotation, annotation_pairs):

    limited_text_graph = networkx.Graph()
    nx_graph = graph.to_undirected()
    print('Undirected networkx graph created.')
    for pair in tqdm_notebook(set(annotation_pairs)):
            
            if pair[0] in nx_graph.nodes() and pair[1] in nx_graph.nodes():
                #print(pair)
                paths = list(networkx.simple_paths.all_simple_paths(nx_graph, 
                                    (pair[0]), 
                                   (pair[1]), 4))
                               
                if paths != []:
                    #print(paths)
                    for path in paths:
                        for related_nodes in list(zip(path, path[1:])):
                            limited_text_graph.add_edge(related_nodes[0], related_nodes[1])
            
    return limited_text_graph 
```

```python
def clean_dictionary(dic):
    '''
    function that cleanes up the dictionary 
    created by the create_dictionary function
    '''
    clean_dic = defaultdict(list)
    if type(dic) == dict:
        for key,val in dic.items():
            for tup in val:
                lst = tup[1].replace('{', '').replace('}','').split('|')
                for item in lst:
                    if 'db' in item:
                        clean_dic[key].append((tup[0], item))
        return clean_dic
```

```python
def turn_to_simple_dic(dic):
    '''
    function that takes the dictionary created by create_dictionary()
    and returns a dictionary excluding the properties 
    in the format  {node: [neighbour]} ie. {subject: [object)]}
    '''
    properties = defaultdict(str)
    simple_dic = defaultdict(str)
    for key in dic:
        if type(dic[key])==list:
            simple_dic[key] = [tup[1] for tup in dic[key] if 'db' in tup[1]]
            for ind,node in enumerate(simple_dic[key]):
                properties[(key,node)] = {'prop': [tup[0] for tup in dic[key] if 'db' in tup[1]][ind]}
    #print(simple_dic, '\n', properties)
    return simple_dic, properties
```

```python
def put_together(dic, annotations):
    dic = re.sub('^[^{]+', '', dic)
    dic = re.sub('[^}]+$', '', dic)
    dic = dic.replace('OrderedDict(', '')
    dic = dic.replace(')])', ')]')
    dic = clean_dictionary(literal_eval(dic))
    clean_dic=turn_to_simple_dic(dic)[0]
    graph = convert.to_networkx_graph(clean_dic)
    #print(turn_to_simple_dic(dic)[1])
    networkx.set_edge_attributes(graph, dict(turn_to_simple_dic(dic)[1]))
    pairs = create_pairs(graph, annotations)    
    text_graph = select_related_nodes(graph, annotations, pairs)
    networkx.set_edge_attributes(text_graph,  dict(turn_to_simple_dic(dic)[1]))
    return text_graph, dict(turn_to_simple_dic(dic)[1])
```
